```python
import pandas as pd

# Path to the large CSV file
file_path = 'test.csv'

# Define the chunk size
chunk_size = 60 # Set a small chunk size for demonstration

# Initialize an empty DataFrame to store the concatenated chunks
result_df = pd.DataFrame()

# Iterate over chunks and process each chunk
for chunk in pd.read_csv(file_path, chunksize=chunk_size):
    # Perform operations on each chunk if needed
    # For example, calculate the average Value for each chunk
    avg_value = chunk['Quantity'].count()

    # Print the result for each chunk
    print(f"Average Value for this chunk: {avg_value}")
    

    # Concatenate the processed chunk to the result DataFrame
    result_df = pd.concat([result_df, chunk], ignore_index=True)

# Now, result_df contains the entire dataset, and you can continue processing the combined data if needed
print("\nCombined DataFrame:")
print(result_df)
```
